```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.experience=df.experience.fillna('zero')
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df.experience
```

```python
df['testscore']=df['test_score(out of 10)']
df['interviewscore']=df['interview_score(out of 10)']
df
```

```python
df
```

```python
df=df.drop('interview_score(out of 10)',axis='columns')
df=df.drop('test_score(out of 10)',axis='columns')

df
```

```python
df
```

```python
ts=df.testscore.mean()
ts
```

```python
import math
ts=math.floor(df['testscore'].mean())
ts
```

```python
df.testscore=df.testscore.fillna(ts)
df
```

```python
reg=linear_model.LinearRegression()
reg
df
```

```python
reg.fit(df[['experience','testscore','interviewscore']],df['salary($)'])
```
