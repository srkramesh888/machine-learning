```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```
